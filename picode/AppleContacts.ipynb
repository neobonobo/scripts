## Importing Apple Contacts

<ul>
    <li>Select all contacts</li>
    <li>Export as vcf to picode folder</li>
</ul>

```python
import vobject
```

```python
infile='Contacts.vcf'
with open(infile) as inf:
    arr=[]
    indata = inf.read()
    vc = vobject.readComponents(indata)
    vo = next(vc, None)
    while vo is not None:
        vo = next(vc, None)
        arr.append(vo)
```

```python
c=1
for card in arr:
    if 'tel' in card.contents.keys():
        print(card.contents['n'][0].value.given,card.contents['n'][0].value.family,card.contents['tel'][0].value)
        
```
